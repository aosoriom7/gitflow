```python
import pandas as pd
import numpy as np

media = 1
stdDev= 1

tam_muestra = 100

df_aleatorio = pd.DataFrame({'va': np.random.normal(loc=media, scale= stdDev, size= tam_muestra)})
```

```python
import matplotlib.pyplot as plt
df_aleatorio.plot(y='va', kind='bar')
plt.xlabel('valor aleat')
plt.ylabel('frec')
plt.title('Aleatorios')
plt.show()
```

```python
plt.hist(df_aleatorio['va'], bins=10)
plt.xlabel('valor aleat')
plt.ylabel('frec')
plt.title('Aleatorios')
plt.show()
```

```python
import plotly.express as px
fig = px.histogram(df_aleatorio, x='va', title='Histograma de Valores')
fig.show()
```
